```python
# save this as domain_encoder.py
import json
import numpy as np
import pandas as pd

def encode_domains(json_data, return_df=True):
    """
    Encode domain data into numerical vectors.
    Returns:
        X (numpy.ndarray): Feature matrix
        y (numpy.ndarray): Target labels
        vocab (dict): Index mapping for skills, keywords, job_titles
        label_map (dict): Domain-to-label mapping
        df (pd.DataFrame, optional): Feature DataFrame with labels
    """

    # ---------- STEP 1: Create label map ----------
    label_map = {domain: idx for idx, domain in enumerate(json_data.keys())}

    # ---------- STEP 2: Collect unique features ----------
    all_skills = set()
    all_keywords = set()
    all_job_titles = set()

    for data in json_data.values():
        all_skills.update(data["skills"])
        all_keywords.update(data["project_keywords"])
        all_job_titles.update(data["job_titles"])

    # Sort for consistent indexing
    all_skills = sorted(list(all_skills))
    all_keywords = sorted(list(all_keywords))
    all_job_titles = sorted(list(all_job_titles))

    # Create index mappings
    vocab = {
        "skills": {skill: idx for idx, skill in enumerate(all_skills)},
        "keywords": {kw: idx for idx, kw in enumerate(all_keywords)},
        "job_titles": {title: idx for idx, title in enumerate(all_job_titles)}
    }

    # Total feature length
    feature_len = len(all_skills) + len(all_keywords) + len(all_job_titles) + 1  # +1 for min_score

    # ---------- STEP 3: Encode each domain ----------
    X = []
    y = []
    domain_names = []

    for domain, data in json_data.items():
        vector = np.zeros(feature_len, dtype=int)

        # Encode skills
        for skill in data["skills"]:
            if skill in vocab["skills"]:
                vector[vocab["skills"][skill]] = 1

        # Encode keywords
        kw_offset = len(all_skills)
        for kw in data["project_keywords"]:
            if kw in vocab["keywords"]:
                vector[kw_offset + vocab["keywords"][kw]] = 1

        # Encode job titles
        jt_offset = len(all_skills) + len(all_keywords)
        for title in data["job_titles"]:
            if title in vocab["job_titles"]:
                vector[jt_offset + vocab["job_titles"][title]] = 1

        # Encode min_score (last index)
        vector[-1] = data["min_score"]

        X.append(vector)
        y.append(label_map[domain])
        domain_names.append(domain)

    X = np.array(X)
    y = np.array(y)

    if return_df:
        # Build DataFrame column names
        columns = (
            list(vocab["skills"].keys())
            + list(vocab["keywords"].keys())
            + list(vocab["job_titles"].keys())
            + ["min_score"]
        )

        df = pd.DataFrame(X, columns=columns)
        df["label"] = y
        df["domain"] = domain_names
        return X, y, vocab, label_map, df

    return X, y, vocab, label_map


# ---------- Example Usage ----------
if __name__ == "__main__":
    # Example: load JSON file
    # with open("domains.json", "r") as f:
    #     json_data = json.load(f)

    X, y, vocab, label_map, df = encode_domains(json_data)

    print("Feature matrix shape:", X.shape)
    print("Target vector shape:", y.shape)
    print("\nSample DataFrame:\n", df.head())
    print("\nLabel mapping:", label_map)
```
